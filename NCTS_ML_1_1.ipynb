```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.io 
from scipy.fftpack import fft
```

```python
data = scipy.io.loadmat('C:/Users/marco/Desktop/IMS6200.mat')
```

```python
IMS = data["IMS"]
ch1 = IMS[0,:]
ch2 = IMS[1,:]
ch3 = IMS[2,:]
ch4 = IMS[3,:]
```

```python
Fs = 20000
T = 1/Fs
L = len(ch1)
t = np.linspace(0,L,L)*T
```

```python
plt.figure(dpi = 1000)
line1 = plt.plot(t, ch1, color = "blue")
line2 = plt.plot(t, ch2, color = "orange")
line3 = plt.plot(t, ch3, color = "green")
line4 = plt.plot(t, ch4, color = "red")
legend = plt.legend(["ch1","ch2","ch3","ch4"])
plt.title("IMS Bearing Data (Time Series)")
plt.xlabel('Time(s)')
plt.ylabel('Gravity(m/s^2)')
plt.grid()
plt.show()
```

```python
f_ch1 = fft(ch1)
f_ch2 = fft(ch2)
f_ch3 = fft(ch3)
f_ch4 = fft(ch4)
xf = np.linspace(0.0, 1.0/(2.0*T), L//2)
```

```python
plt.figure(dpi = 1000)
f_1 = plt.plot(xf, 2.0/L*np.abs(f_ch1[0:L//2]))
f_2 = plt.plot(xf, 2.0/L*np.abs(f_ch2[0:L//2]))
f_3 = plt.plot(xf, 2.0/L*np.abs(f_ch3[0:L//2]))
f_4 = plt.plot(xf, 2.0/L*np.abs(f_ch4[0:L//2]))
legend = plt.legend(["ch1","ch2","ch3","ch4"])
plt.title("IMS Bearing Data (Fourier Transform)")
plt.xlabel('Frequency(Hz)')
plt.ylabel('Gravity(m/s^2)')
plt.grid()
plt.show()
```
