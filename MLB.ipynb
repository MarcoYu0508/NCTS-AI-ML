```python
import seaborn as sns
%matplotlib inline 
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('2018 Batters Leaderboard.csv', index_col="Name")
```

```python
df
```

```python
df = df[df['WAR']>0]
```

```python
df
```

```python
bins = [0,1,2,3,4,5,6,15]
bins
```

```python
group_names = ['Scurb', 'Role Player', 'Solid Starter', 'Good Player', 'All_Star','SuperStar', 'MVP']
group_values = [0,1,2,3,4,5,6]
```

```python
df['WAR-bins'] = pd.cut(df['WAR'], bins, labels =group_names, include_lowest=True)
df['Values'] = pd.cut(df['WAR'], bins, labels =group_values, include_lowest=True)
df[['WAR','WAR-bins','Values']]
```
