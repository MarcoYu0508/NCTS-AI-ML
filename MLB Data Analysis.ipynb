```python
%matplotlib inline 
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import seaborn as sns
```

資料來源:https://www.fangraphs.com/leaders.aspx?pos=all&stats=bat&lg=all&qual=y&type=8&season=2018&month=0&season1=2018&ind=0&team=0&rost=0&age=&filter=&players=0

```python
p_df = pd.read_csv('2018 Pitchers Leaderboard.csv', index_col="Name")#投手的資料(投球局數大於50局)
b_df = pd.read_csv('2018 Batters Leaderboard.csv', index_col="Name")#打者的資料(打席數大於100)
```

```python
p_df.head(10)
```

```python
b_df.head(10)
```

```python
#取出WAR>0的資料
#WAR全名Wins Above Replacement，WAR指的是球員的綜合評價，包含進攻跟防守在內，且投、打皆可以計算出的數據。
p_df2 = p_df[p_df['WAR']>0]
b_df2 = b_df[b_df['WAR']>0]
```

```python
#用WAR分級， MVP(6+ WAR)、SuperStar(5~6WAR)、All_Star(4~5WAR)、Good Player(3~4WAR)、Solid Starter(2~3WAR)、Role Player(1~2WAR)、Scurb(0~1WAR)
bins = [0,1,2,3,4,5,6,15]
group_names = ['Scurb', 'Role Player', 'Solid Starter', 'Good Player', 'All_Star','SuperStar', 'MVP']
group_values = [0,1,2,3,4,5,6]
```

```python
b_df2['WAR-bins'] = pd.cut(b_df2['WAR'], bins, labels =group_names, include_lowest=True)
b_df2[['WAR','WAR-bins']].head(10)
```

```python
p_df2['WAR-bins'] = pd.cut(p_df2['WAR'], bins, labels =group_names, include_lowest=True)
p_df2[['WAR','WAR-bins']].head(10)
```

```python
def plotComparePitcherWAR():
    #不同等級投手與 ERA(防禦率)的散佈情形
    plt.figure(figsize=(8,4))
    sns.stripplot(x="WAR-bins", y="ERA", data=p_df2)
    plt.title("Pitcher")
    plt.grid()
    plt.show()
```

```python
def plotCompareBatterWAR():
    #不同等級投手與 OBP(上壘率)的散佈情形
    plt.figure(figsize=(8,4))
    sns.stripplot(x="WAR-bins", y="OBP", data=b_df2)
    plt.title("Batter")
    plt.grid()
    plt.show()
```

```python
def plotPitcherWAR():
    #每隊所擁有的投手WAR的分布情形
    plt.figure(dpi=100)
    ax = sns.stripplot(x="Team", y="WAR", data=p_df)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
    plt.title("Pitcher")
    plt.grid()
    plt.show()
```

```python
def plotBatterWAR():
    #每隊所擁有的打者WAR的分布情形
    plt.figure(dpi=100)
    ax = sns.stripplot(x="Team", y="WAR", data=b_df)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
    plt.title("Batter")
    plt.grid()
    plt.show()
```

```python
def findTeamWARbins():
    team_name = input("Please input your team name ")
    team_p = p_df2[p_df2['Team']==team_name]
    team_b = b_df2[b_df2['Team']==team_name]
    print(f"Player for {team_name}")
    print('Pitcher')
    display(team_p['WAR-bins'].value_counts().to_frame())
    print('Batter')
    display(team_b['WAR-bins'].value_counts().to_frame())
```

```python
def select():
    selection = int(input('''please input your selcetion:
                            0. Close
                            1. Compare Pitcher WAR with ERA
                            2. Compare Batter WAR with OBP
                            3. Pitcher WAR
                            4. Batter WAR
                            5. TeamWAR\n'''))

    while selection != 0:
        if selection == 1:
            plotComparePitcherWAR()
        elif selection == 2:
            plotCompareBatterWAR()
        elif selection == 3:
            plotPitcherWAR()
        elif selection == 4:
            plotBatterWAR()
        elif selection == 5:
            findTeamWARbins()
        else:
            print('Error Input')
        selection = int(input('''please input your selcetion:
                            0. Close
                            1. ComparePitcherWAR
                            2. CompareBatterWAR
                            3. PitcherWAR
                            4. BatterWAR
                            5. TeamWAR\n'''))
```
